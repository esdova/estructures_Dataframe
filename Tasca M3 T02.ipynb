```python
import numpy as np
import pandas as pd     #0k vols enderrerits

df = pd.read_csv ( 'DelayedFlights.csv',index_col=0)      #creo una columna nova totalDelay amb la suma  
df = df.sample(n = 200000)                                #de tots els edarreriments( vull dir. sumo  tots els valors 
                                                          # de les columnes que contenen retards) amb np.nan_to_num, 
                                                          #puc convertir en 0 el valor NaN

df["totalDelay"] = np.nan_to_num(df['ArrDelay']) + np.nan_to_num(df['DepDelay']) + np.nan_to_num(df['CarrierDelay']) + np.nan_to_num(df['WeatherDelay']) + np.nan_to_num(df['NASDelay']) + np.nan_to_num(df['SecurityDelay']) + np.nan_to_num(df['LateAircraftDelay'])

print(df[["UniqueCarrier","ArrDelay","totalDelay"]])

agrupado = df.groupby(['UniqueCarrier']).agg({'totalDelay': 'sum'}).sort_values(by='totalDelay', ascending=False);

print(agrupado)    # agrupo les companyies 'UniqueCarrier'amb groupby   i, amb    totalDelay obtinc el total de les sumes
                   # dels endarreriments 
                   # sort_values els ordena de més gran a mes petit
```
